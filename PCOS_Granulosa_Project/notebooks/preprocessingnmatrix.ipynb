```python
import pandas as pd
files = {"Ctrl_1": r"C:\Users\Public\Documents\bioinformatics-gene-expression\data\GSM8898319_Ctrl-1.polya_influence.xls",
    "Ctrl_2": r"C:\Users\Public\Documents\bioinformatics-gene-expression\data\GSM8898320_Ctrl-2.polya_influence.xls",
    "Ctrl_3": r"C:\Users\Public\Documents\bioinformatics-gene-expression\data\GSM8898321_Ctrl-3.polya_influence.xls",
    "PCOS_1": r"C:\Users\Public\Documents\bioinformatics-gene-expression\data\GSM8898322_PCOS-1.polya_influence.xls",
    "PCOS_2": r"C:\Users\Public\Documents\bioinformatics-gene-expression\data\GSM8898323_PCOS-2.polya_influence.xls",
    "PCOS_3": r"C:\Users\Public\Documents\bioinformatics-gene-expression\data\GSM8898324_PCOS-3.polya_influence.xls",
}

dfs = []
for sample, path in files.items():
    df = pd.read_csv(path, sep="\t")
    df = df[["gene", "tpm"]].set_index("gene")
    df.columns = [sample]
    dfs.append(df)

expr = pd.concat(dfs, axis = 1)
expr = expr.fillna(0)

expr.head()
```
